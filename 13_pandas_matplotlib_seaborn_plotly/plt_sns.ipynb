створюємо датафрейм та док в ексель

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.DataFrame(
    {
        'mark': ['q','w','e','r','t','y'],
        "price":[1000,1200,900,5000,1300,3000],
        'value':[8,7,5,10,7,6],
        'speed':[150,170,140,300,165,220],
        'carier':[1,0.8,2,0.6,1.2,2.1]
    }
)
df.to_excel('F:/data_science_course/git/Lesson_33/DF.xlsx', sheet_name='auto', index=False)
print(df)
```

Лінійна діаграма

```python
auto= pd.read_excel('F:/data_science_course/git/Lesson_33/DF.xlsx')
plt.plot(auto.value,auto.price)
```

Точкові діаграми

```python
plt.scatter(auto.value,auto.speed, color='black')
```

```python
plt.scatter(auto.carier,auto.speed, color='red',linewidth=10)
```

Гістограмма

```python
plt.hist(auto.speed, 5, color='violet' )
```

Коробковий графік

```python
sns.boxplot(x='mark', y='speed', data=auto)
```
